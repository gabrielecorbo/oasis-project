```python
import pandas as pd
big_df = pd.read_csv("utd19_u.csv")
```

```python
big_df
```

```python
import numpy as np
cities = np.unique(list(big_df["city"]))
```

```python
cities
```

```python
coords = pd.read_csv("detectors_public.csv")
coords
```

```python
big_df['detid'] = big_df['detid'].astype(object)
coords['detid'] = coords['detid'].astype(object)
```

```python
df = big_df.merge(coords, on="detid", how = "left")
df
```

```python
for city_name in cities:
    df_city = df[df["city"]==city_name]
    df_city = df_city.groupby(by=["detid"]).mean()
    df_city.to_csv(city_name+"_traffic_points.csv")
```
