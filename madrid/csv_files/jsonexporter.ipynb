```python
import os
import pandas as pd
import json
from pandas import json_normalize

directory = os.fsencode(os.getcwd() + '\\ES')
```

```python
result = pd.DataFrame()
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".asm") or filename.endswith(".json"): 
        #print(file)
        #print(os.path.join(directory, file))
        f = open(os.path.join(directory, file),encoding="utf8")
        # returns JSON object as 
        # a dictionary
        data = json.load(f)
        #data = json.load(os.path.join(directory, file))
        #print(data)
        #print(data['AddressInfo'])
        # Use pandas.DataFrame.from_dict() to Convert JSON to DataFrame
        df2 = pd.DataFrame.from_dict(data['AddressInfo'], orient="index")
        result = pd.concat([result,df2.transpose()])
        continue
    else:
        continue

print(result)
```

```python
df = result[result['Town']=='Madrid']
print(df)
```

```python
from shapely.geometry import Point
import geopandas as gpd
def point_df_to_gdf(df):
    """takes a dataframe with columns named 'longitude' and 'latitude'
    to transform to a geodataframe with point features"""

    df['coordinates'] = df[['Longitude', 'Latitude']].values.tolist()
    df['coordinates'] = df['coordinates'].apply(Point)
    df = gpd.GeoDataFrame(df, geometry='coordinates')
    return df

gdf = point_df_to_gdf(df)
gdf.to_file(os.getcwd()+'\\provashape\\EV_points.shp')
```
